```python
import pandas as pd
import numpy as np
from lifelines import CoxPHFitter
from lifelines import KaplanMeierFitter
from lifelines import WeibullAFTFitter
import matplotlib.pyplot as plt
```

### Variable info
1. ID: unique identifier
2. N_Days: number of days between registration and the earlier of death, transplantation, or study analysis time in July 1986
3. Status: status of the patient C (censored), CL (censored due to liver tx), or D (death)
4. Drug: type of drug D-penicillamine or placebo
5. Age: age in [days]
6. Sex: M (male) or F (female)
7. Ascites: presence of ascites N (No) or Y (Yes)
8. Hepatomegaly: presence of hepatomegaly N (No) or Y (Yes)
9. Spiders: presence of spiders N (No) or Y (Yes)
10. Edema: presence of edema N (no edema and no diuretic therapy for edema), S (edema present without diuretics, or edema resolved by diuretics), or Y (edema despite diuretic therapy)
11. Bilirubin: serum bilirubin in [mg/dl]
12. Cholesterol: serum cholesterol in [mg/dl]
13. Albumin: albumin in [gm/dl]
14. Copper: urine copper in [ug/day]
15. Alk_Phos: alkaline phosphatase in [U/liter]
16. SGOT: SGOT in [U/ml]
17. Triglycerides: triglicerides in [mg/dl]
18. Platelets: platelets per cubic [ml/1000]
19. Prothrombin: prothrombin time in seconds [s]
20. Stage: histologic stage of disease (1, 2, 3, or 4)

Class Labels

Status: status of the patient 0 = D (death), 1 = C (censored), 2 = CL (censored due to liver transplantation)

```python
df = pd.read_csv('cirrhosis.csv') #Read the data into a pandas dataframe
```

## Pre-processing and label encoding
Steps taken:


-Observations with na values are dropped 
<br>
-'event_observed' variable is created for later KM/CoxPH fitting based on value in 'Status' variable: 1 for value of D, 0 for C or CL
<br>
-'Edema' variable is label encoded: Y=2, S=1, N=0
<br>
-'Sex','Ascites','Hepatomegaly','Spiders','Drug' are all label encoded as 1 for value of Female, Y, or if the subject took D-penicillamine respectively, 0 else
<br>
-Extra variables that were created by pd.get_dummies() are then dropped, as well as 'Edema' 'ID 'Status' as the information was encoded numerically

```python
for _ in df.columns:
    if sum(df[_].isna()) > 0:
        print(_)
    else:
        continue
```

```python
df = df.dropna()
df['event_observed'] = df['Status'].apply(lambda x: 1 if x == 'D' else 0)
df["Edema"] = df["Edema"].astype('category')
df['Edema_cat'] = df['Edema'].cat.codes #Y=2 S=1 N=0
df = pd.get_dummies(df, columns=['Sex','Ascites','Hepatomegaly','Spiders','Drug'])
df=df.drop(labels=['Edema','ID','Status','Sex_M','Ascites_N','Hepatomegaly_N','Spiders_N','Drug_Placebo'], axis=1)
df.head()
```

```python
# We create a variable to later separate observations where D-penicillamine was taken
strata_var = df['Drug_D-penicillamine'] == 1 
```

```python
ax = plt.subplot(111)

# Create the Kaplan-Meier fiftter object
kmf = KaplanMeierFitter()
# Fit the stratified data to the kmf object
kmf.fit(durations=df[~strata_var]['N_Days']/360, event_observed=df[~strata_var]['event_observed'],label='Placebo')
# Plot curve 1 'Placebo'
kmf.plot_survival_function(ax=ax)

kmf.fit(durations=df[strata_var]['N_Days']/360, event_observed=df[strata_var]['event_observed'],label='Drug')
# Plot curve 2 'Drug'
kmf.plot_survival_function(ax=ax)
plt.title('Survival functions for Placebo vs D-penicillamine groups')
```

```python
# Create Cox Proportional Hazards model
cph = CoxPHFitter()
# Fit CPH to data (this is not by strata)
cph.fit(df, duration_col='N_Days', event_col='event_observed')

cph.print_summary() 
```

```python
# Backward selection stuff
threshold = 0.05
while True:
    cph.fit(df, duration_col='N_Days', event_col='event_observed')
    pvals = cph.summary['p']
    max_p = pvals.max()
    if max_p > threshold:
        worst_feature = pvals.idxmax()
        print(f"Dropping '{worst_feature}' (p = {max_p:.4f})")
        df = df.drop(columns=[worst_feature])
    else:
        break
```

## Gamma Regression

```python
df_gamma = df
```

```python
df_gamma = df_gamma.rename(columns={'Drug_D-penicillamine': 'Drug_D_penicillamine'})
```

```python
df_gamma.columns
```

```python
import statsmodels.formula.api as smf
import statsmodels.api as sm
#full model test
model = smf.glm(
    formula='N_Days ~ Age + Bilirubin + Cholesterol + Albumin + Copper + Alk_Phos + SGOT + Tryglicerides + Platelets + Prothrombin + Stage + Edema_cat + Sex_F + Ascites_Y + Hepatomegaly_Y + Spiders_Y + Drug_D_penicillamine',
    data=df_gamma,
    family=sm.families.Gamma(sm.families.links.log())
)

result = model.fit()
print(result.summary())

```

```python
import statsmodels.formula.api as smf

predictors = [
    'Age', 'Bilirubin', 'Cholesterol', 'Albumin', 'Copper',
    'Alk_Phos', 'SGOT', 'Tryglicerides', 'Platelets', 'Prothrombin',
    'Stage', 'Edema_cat', 'Sex_F', 'Ascites_Y', 'Hepatomegaly_Y',
    'Spiders_Y', 'Drug_D_penicillamine'
]

response = 'N_Days'
threshold = 0.05  # Sig thresh

# more backward selection stuffs
while True:
    # Fit model
    formula = f"{response} ~ {' + '.join(predictors)}"
    model = smf.glm(
        formula=formula,
        data=df_gamma,
        family=sm.families.Gamma(sm.families.links.log())
    ).fit()
    
    #Get p-values for all terms
    pvals = model.pvalues.drop('Intercept')
    
    # Find the predictor with the highest p-value
    max_pval = pvals.max()
    worst_term = pvals.idxmax()

    # Check if significant
    if max_pval < threshold:
        print("All remaining predictors are significant.")
        break

    # Remove the predictor 
    # Remove base variable name from the predictor list
    base_name = worst_term.split('[')[0]  
    print(f"Dropping '{worst_term}' (p = {max_pval:.4f})")
    predictors = [p for p in predictors if not p.startswith(base_name)]

print("\nFinal model formula:")
print(f"{response} ~ {' + '.join(predictors)}")
print(model.summary())

```

```python
import matplotlib.pyplot as plt

# Coefficients and CIs
variables = ['Bilirubin', 'Albumin', 'Copper', 'Alk_Phos', 'Stage', 'Edema_cat']
coef = [-0.0398, 0.4034, -0.0015, 5.858e-05, -0.1014, -0.2575]
lower = [-0.056, 0.236, -0.002, 2.96e-05, -0.178, -0.387]
upper = [-0.024, 0.571, -0.001, 8.76e-05, -0.025, -0.128]

plt.figure(figsize=(8, 5))
y_pos = range(len(variables))

# Plot with error bars
plt.errorbar(coef, y_pos, xerr=[[c - l for c, l in zip(coef, lower)],
                                [u - c for u, c in zip(upper, coef)]],
             fmt='o', color='steelblue', ecolor='gray', capsize=5)

plt.yticks(y_pos, variables)
plt.axvline(0, color='black', linestyle='--')
plt.xlabel("Coefficient (log-scale effect)")
plt.title("Gamma Regression Coefficients with 95% CI")
plt.tight_layout()
plt.show()
```
